### Set data pipelines 

```python
def extract_num(df, target):
    
    num = list(df.select_dtypes(include="number").columns)
    if target in num:
        num.remove(target)

    return num

def extract_cat(df, target):
    
    cat = list(df.select_dtypes(exclude="number").columns)
    if target in cat:
        cat.remove(target)
        
    return cat

def buildpipe(name, model, num, cat):
    
    # sub pipeline numeric
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='mean')),
        ('scaler', StandardScaler())])
        
    # sub pipeline categorical
    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))])

    # map data types to their transformers
    preprocessor = ColumnTransformer(
        transformers=[
            ("num", numeric_transformer, num),
            ("cat", categorical_transformer, cat)])

    # define main pipe main pipe
    mainpipe = Pipeline(steps=[
        ('preprocessor', preprocessor), (name, model)])

    return mainpipe, preprocessor
```

```python
# check scaling effects

num, cat = extract_num(df, target), extract_cat(df, target)

_ , preprocessor = buildpipe(name, model, num, cat)

x_scaled = preprocessor.fit_transform(df.drop(columns=target))
        
dfx_scaled = pd.DataFrame(x_scaled, index=df.index, columns=df.drop(columns=target).columns)
        
# boxplot
dfx_scaled.boxplot(figsize=(10,10))
plt.show()
```

### Train 

```python
def train(df_clean, target, model, name, param, shuffle=True):
    
    X, y = df_clean.drop(columns=target), df_clean[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, shuffle=shuffle)
    
    num, cat = extract_num(df_clean, target), extract_cat(df_clean, target)
    
    if param is not None:
        
        keys = list(param)
        if type(param[keys[0]]) == list:
            grid_search = True
        else:
            grid_search = False
    
    if param is None:
        grid_search = False
        
    if grid_search:
        print("grid search " + name + "...")

        mainpipe, _ = buildpipe(name, model, num, cat)
                
        # create search object
        param_grid = {name+'__'+k:v for k, v in param.items()}
    
        grid = GridSearchCV(mainpipe,
                          param_grid=param_grid,
                          scoring='accuracy',
                          cv=None,
                          n_jobs=-1)

        # fit data
        model = grid.fit(X_train, y_train)
        
    if grid_search==False:
        print("fitting " + name + " with param : ", param)
        if param is not None:
            model = model.set_params(**param)

        mainpipe, _ = buildpipe(name, model, num, cat)

        # fit data
        model = mainpipe.fit(X_train, y_train)
    
    # predict test data (and train data to check overfitting)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # calculate metrics
    if grid_search:
        print(name + " best param :" + str(model.best_params_))
    
    print('acc. train :' + str(accuracy_score(y_train, y_train_pred)))
    print('acc. test :' + str(accuracy_score(y_test, y_test_pred)))
        
    #Confusion matrix
    cm = confusion_matrix(y_test, y_test_pred)
    cmp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y_train))
    fig, ax = plt.subplots(figsize=(10,10))
    cmp.plot(cmap=plt.cm.Blues, ax=ax)    
    plt.title(name)

    return model
```

### Run  

```python
param_grid_knn = {'n_neighbors':[3, 7, 21, 31]}
param_grid_svm = {'C': [0.1, 100], 'gamma': [0.01, 1]}
param_grid_tree = {'max_depth': [2, 6, 12]}
param_grid_lr = {"C":[0.01,1,100], "penalty":["l2", "none"]}
param_grid_rf = {'max_depth': [3,5,10],
           'min_samples_leaf': [1, 2, 4],
           'min_samples_split': [2, 5, 10],
           'n_estimators': [64, 256, 512]}

model_select = {
    "knn":KNeighborsClassifier(),
    "svm":SVC(kernel='rbf'),
    "tree":DecisionTreeClassifier(criterion="gini"),
    "naive bayes":GaussianNB(),
    "lr":LogisticRegression(solver="lbfgs", max_iter=1500),
    "rf": RandomForestClassifier(),
}

param_grid = {
    "knn":param_grid_knn,
    "svm":param_grid_svm,
    "tree":param_grid_tree,
    "naive bayes":None,
    "lr":param_grid_lr,
    "rf":param_grid_rf,

}
```

```python
name = "knn"
model = model_select[name]
param = param_grid[name]

model = train(df, target, model, name, param, shuffle=True)
```
